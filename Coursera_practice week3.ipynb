```python
df= pd.read_csv(r"C:\Users\Юрий\Desktop\Занятия\Coursera\resources\week-3\datasets\census.csv")
df=df[df['SUMLEV']==50]
df.head()
```

```python
%%timeit -n 3
# по каждому штату выводим среднее значение 'CENSUS2010POP' через цикл for
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in state ' + state + 
         'have an average population of' + str(avg))
```

```python
%%timeit -n 3
#теперь делаем то же самое но через group by
#group - это группы строк разделенные по имени штата.
#frame - это датафрейм каждой группы. 
# в цикле ищем среднее столбца 'CENSUS2010POP' по каждому фрейму каждой группы (каждого штата)
for group, frame in df.groupby('STNAME'): 
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group +
         ' have an average population of ' + str(avg))
```

```python
#способ с groupby быстрее почти в 100 раз.
```

# Transformation

```python
# Transformation - отличается от agg, тем что возвращает не одну строку на группу, а фрейм того же размера что и группа. 
# по сути от транслирует функцию по всему датафрейму и возвращает новый датафрейм.
```

# Applying

```python
# самая популярная функция для groupby это apply().
# позволяет применить произвольную функцию к каждой группе и сшить результаты обратно в единый датафрейм. 
# напр:
df=pd.read_csv(r"C:\Users\Юрий\Desktop\Занятия\Coursera\resources\week-3\datasets\listings.csv")
df=df[['cancellation_policy', 'review_scores_value']]
df.head()
```

```python
#то что мы делали в transform, в aplly можно сделать проще, написав доп.функцию 
def calc_mean_review_scores(group):
    # Функция будет работать сразу с целым датафреймом группы
    avg=np.nanmean(group['review_scores_value'])
    #далее транслируем нашу формулу и создадим новую группу
    group['review_scores_mean']=np.abs(avg-group['review_scores_value'])
    return group
df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```

```python
#использование apply может быть медленнее специализированных функций, особенно agg
#но если датафреймы не огромные то это надежный подход общего назначения. 
```

# Pivot Table (Сводная таблица)

```python
import pandas as pd
import numpy as np
# Pivot table - является способом суммирования данных, где строки представляют одну переменную, 
# столбцы другую, а ячейки некоторое совокупное значение. Также это могут быть предельные значения являющиеся суммой для каждого столбца и строки. 
# такая таблица позволяет нам видеть связь между двумя переменными с первого взгляда.
df = pd.read_csv(r"C:\Users\Юрий\Desktop\Занятия\Coursera\resources\week-3\datasets\cwurData.csv")
df # Датафрейм предсатвляет рейтинг университетов
```

```python
# предположим мы хотим создать новую колонку Rank_level, в которой университеты с рейтингом 1-100 классифицируются как первый уровень,
# 101-200, а 201-300 третий уровень. после 300 университеты без ранга. 
def create_category(ranking):
    if (ranking>=1)&(ranking<=100):
        return 'First Tier Top University'
    elif (ranking>=101)&(ranking<=200):
        return 'Second Tier Top University'
    elif (ranking>=201)&(ranking<=300):
        return 'Third Tier Top University'
    return 'Other Top University'

df['Rank_level'] = df['world_rank'].apply(lambda x:create_category(x))
df.head()
```

```python
# далее мы хотим сравнить уровень ранга с страной университета, поэтому сравниваем по общему баллу. 
# поэтому мы делаем индексом страну, значения - scores, столбцы уровни ранга. и найдем средний ранг для страны. 
df.pivot_table(values='score', index='country',columns='Rank_level', aggfunc=[np.mean]).head()
```

# Date/Time Functionality

```python
#рассмотрим временные ряды и функциональные возможности дат в Пандас
#т.к. пандас был разработан для обработки транзакций хедж-фондов, то он имеет гибкую систему работы с датами и временем. 
```

```python
import pandas as pd
import numpy as np
```

# Timestamp

```python
# в пандас есть четыре основных класса для времени:
# Timestamp
# DatatimeIndex
# Period
# PeriodIndex
```

```python
# Timestamp (временная метка)
# создадим метку со строкой 9/1/2019 10:05AM,
# в большинстве случаев Timestamp взаимозаменяем с Datetime в Python
pd.Timestamp('9/1/2019 10:05AM')
```

```python
# можно передать данные в другом формате. 
pd.Timestamp(2019, 12, 20, 0,0)
```

```python
# также у Timestamp есть полезные атрибуты - isoweekday - показывает день недели. 
pd.Timestamp(2019, 12, 20, 0,0).isoweekday() # 1 - понедельник, 7 - воскресенье
```

```python
# также можно извлечь из timestamp секунды через атрибут second
pd.Timestamp(2019, 12, 20, 5, 2, 12).second
```

# Period

```python
# Period - показывает промежуток времени
```

```python
pd.Period('1/2016') # январь
```

```python
pd.Period('3/5/2016') #пятое марта 
```

```python
# хотим узнать 5 месяцев после января 2016го
pd.Period('1/2016') + 5
```

```python
pd.Period('3/5/2016') - 2
```

```python
t2 = pd.Series(list('def'), [pd.Period('2016-09'),pd.Period('2016-10'),
                             pd.Period('2016-11')])
t2
```

# Converting to Datetime

```python
d1 = ['2 June 2013', 'Aug 29, 2014', '2015-06-26', '7/12/16']
ts3 = pd.DataFrame(np.random.randint(10, 100,(4,2)), index=d1,
                  columns=list('ab'))
ts3
```

# Timedelta

```python
# разница во времени. 
pd.Timestamp('9/3/2016')-pd.Timestamp('9/1/2016')
```

```python
pd.Timestamp('9/2/2016 8:10AM') + pd.Timedelta('12D 3H') 
```

# Offset

```python
#Offset похоже на Timedelta но следует определенным правилам продолжительности календаря
# помимо часа, дня, недели, месяца, включает в себя рабочий день, конец месяца, пол месяца и т.д. 
pd.Timestamp('9/4/2016').weekday()
```

```python
pd.Timestamp('9/4/2016') + pd.offsets.Week()
```

```python
pd.Timestamp('9/4/2016') + pd.offsets.MonthEnd()
```

# Working with Dates in a DataFrame

```python
# предположим мы хотим посмотреть на 9 измерений проводимых раз в две недели, каждое воскресенье, начиная с октября 2016го
# 2W-SUN - раз в две недели в воскресенье. 
dates = pd.date_range('10-01-2016', periods=9, freq='2W-SUN')
dates
```

```python
# c помощью этой же функции можем посмотреть рабочие дни (9 рабочих дней с определенной даты)
pd.date_range('10-01-2016', periods=9, freq='B')
```

```python
# ежеквартально начиная с июня. 
pd.date_range('04-01-2016', periods=12, freq='QS-JUN')
```

```python
dates = pd.date_range('10-01-2016', periods=9, freq='2W-SUN')
df = pd.DataFrame({'Count 1': 100 + np.random.randint(-5, 10, 9).cumsum(),
                  'Count 2': 120 + np.random.randint(-5, 10, 9)}, index=dates)
df
```

```python
# можем проверить в какой день недели есть определенные данные
df.index.weekday
```

```python
# можем использовать diff чтобы найти разницу между датами
df.diff()
```

```python
# среднее количество для каждого месяца в df
df.resample('M').mean()
```

```python
# посмотрим на индексацию дат.
# если хотим 2017й год то
df['2017']
```

```python
# если хотим определенный месяц 
df['2016-12']
```

```python
# можем поставить интервал дат (слайс)
df['2016-12':]
```
