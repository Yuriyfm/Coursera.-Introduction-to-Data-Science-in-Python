```python
import pandas as pd
import numpy as np

Energy = pd.read_excel(r"C:\Users\Юрий\Desktop\Занятия\Coursera\assignments\assignment3\assets\Energy Indicators.xls")
Energy = Energy[(Energy.index>14)&(Energy.index<244)]
Energy.drop(Energy.columns[[0, 1]], axis=1, inplace=True)
Energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
Energy.drop(Energy.index[[0,1]], inplace=True)
# почистили исходный файл. убрали заголовок (верхние 14 строк и примечание, нижние строки после 244й)
# убрали первые две колонки(пустые), колонки переименовали как требуется в задании.удалили две верхние строки (пустые).
Energy
```

```python
cols =['Energy Supply', 'Energy Supply per Capita', '% Renewable']
mask = Energy[cols].applymap(lambda x: isinstance(x, (int, float)))
Energy[cols] =  Energy[cols].where(mask)
Energy.loc[:,'Energy Supply'] *= 1000000
Energy
# заменили все нечисловые значения на NAN. И умножили Energy Supply на 1 000 0000 (перевели петаджоули в гигаджоули). теперь везде гигаджоули.
```

```python
pattern = ("\d.*| \(.+")  #X(?=Y)
Energy['Country'].replace((pattern),'',regex=True, inplace=True)
Energy = Energy.set_index('Country')
Energy = Energy.rename(index={"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"})
Energy.head(50)
#with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    #print(Energy)
```

# Question 10
Create a new column with a 1 if the country's % Renewable value is at or above the median for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

This function should return a series named HighRenew whose index is the country name sorted in ascending order of rank.

```python
result['med'] = result['% Renewable'].median()
def f_med(Renewable, med):
        if Renewable>=med:
            return 1
        elif Renewable<med:
            return 0
result['% Renewable_med'] = result.apply(lambda x: f_med(x['% Renewable'], x['med']), axis=1)
HighRenew =result['% Renewable_med'][::-1]
result['PopEst'].std()
```
