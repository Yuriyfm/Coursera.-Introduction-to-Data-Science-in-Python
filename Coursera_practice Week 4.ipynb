```python
import numpy as np
import pandas as pd
from scipy import stats
# scipy - коллекция библиотек для data science. она включает в себя numpy и pandas
# а также библиотеки для визуализации как matplotlib и др. 
```

```python
# у нас есть нулевая гипотеза, и некая альтернативная гипотеза которую мы хотим подтвердить
# рассмотрим пример: 
df = pd.read_csv (r"C:\Users\Юрий\Desktop\Занятия\Coursera\resources\week-4\datasets\grades.csv")
df.head()
```

```python
# разделим студентов на две части. первая, те кто завершил первое задание к концу декабря 2015г. 
# вторая - те кто позже
early_finishers = df[pd.to_datetime(df['assignment1_submission'])<'2016']
late_finishers=df[~df.index.isin(early_finishers.index)]
# ~ - обозначение всего что не  входит в указанный датафрейм (отрицание)
# т.е. только те индексы в df что не early_finishers.index
# это довольно сложный способ, можно решить проще.. 
```

```python
# теперь рассмотрим статистические функции 
# сравним средние значения для двух наших групп. 
print(early_finishers['assignment1_grade'].mean())
print(late_finishers['assignment1_grade'].mean())
```

```python
# проверим насколько значимы эти различия (влияет ли время быполнения первого эказамена, на итоговый результат)
# порогом будет 0.05 (порог занчимости обычно называют Альфа (??))
# scipy содержит ряд статистических тестов и служит основой для статистических гипотез в питоне
# будем использовать функцию ttest_ind(), которая делает independent T-test(проверяем что группы не связаны друг с другом. )
# резултатом T-test является t_statistic and p-value.

from scipy.stats import ttest_ind
ttest_ind(early_finishers['assignment1_grade'], late_finishers['assignment1_grade'])
```

```python
# видим что Р = 0.18, это выше нашего значения Альфа 0.05
# значит мы не можем отклонить нулевую гипотезу. т.е. результат не зависит от времени сдачи (вероятно)

# теперь проверим другие оценки 
print(ttest_ind(early_finishers['assignment2_grade'], late_finishers['assignment2_grade']))
print(ttest_ind(early_finishers['assignment3_grade'], late_finishers['assignment3_grade']))
print(ttest_ind(early_finishers['assignment4_grade'], late_finishers['assignment4_grade']))
print(ttest_ind(early_finishers['assignment5_grade'], late_finishers['assignment5_grade']))
print(ttest_ind(early_finishers['assignment6_grade'], late_finishers['assignment6_grade']))
```

```python
# видим что по всем экзаменам нет значимых результатов. 
# данные значения P - могут помочь нам в дальнейшем экспериментальном дизайне
# напр assignment3_grade имеет P около 0.1 и если бы мы установили АЛьфу в 11% то результат был бы значимым.
# это подсказывает нам на что стоит смотреть в дальнейшем. если бы у нас было небольшое количество участников 
# или в этом задании было бы что-то уникальное, то мы могли бы исследовать глубже в ту сторону чтобы попытаться найти значимые результаты. 
# 
```

```python
# сейчас P-values подвергают критике, за то что они недостаточны чтобы рассказать нам о "взаимодействиях".
# и два других метода - доверительные интервалы, и байесовский анализ используют чаще.
# одна из проблем  P-values закл. в том что если делать много разных тестов то рано или поздно мы получим значимый результат. 
# давайте посмотрим на такую симуляцию. 

# создадим датафрейм из 100 столбцов по 100 строк. 
df1= pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
df2= pd.DataFrame([np.random.random(100) for x in range(100)])
df2.head()
```

```python
# мы точно знаем что эти фреймы рандомны и зависимостей внути нет. 
# возьмем альфу в 10%. будем сравнивать все столбцы в фреймах попарно. 
# оставим только результаты где Р >10%
def test_columns(alpha=0.1):
    num_diff=0
    for col in df1.columns:
        teststat,pval=ttest_ind(df1[col],df2[col])
        if pval<=alpha:
            print('Col {} is statistifically significantly different at alpha={}, pval={}'.format(col,alpha,pval))
            num_diff+=1
    print('Total number different was {}, which is {}%'.format(num_diff,float(num_diff)/len(df1.columns)*100))
test_columns()
```

```python
# итак получили 13 значимых результатов из 100, в рандомных фреймах. 
# попробуем с альфой 0.05
test_columns(0.05)
```

```python
# видим что T-test может ошибаться и очень важно устанавливать правильный порог Альфа. 
# для установления порога желательно пообщаться с экспертами в нужной области и понять что они считают значимым. 
#  
```
