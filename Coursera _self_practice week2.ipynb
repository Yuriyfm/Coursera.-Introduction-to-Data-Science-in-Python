```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv(r"C:\Users\Юрий\Desktop\Занятия\Coursera\resources\week-2\datasets\presidents.csv")
df
```

```python
df=df.set_index("#")
df.head(10)
```

```python
#разбиваем колонку президент на две(имя и фамилию), через apply 
def Splitname(row):
    row['First']=row['President'].split(" ")[0]
    row['Last']=row['President'].split(" ")[-1]
    return row
df=df.apply(Splitname, axis='columns')
del df['President']
df.head(10)
```

```python
df=df [['Last','First','Born','Age atstart of presidency','Age atend of presidency','Post-presidencytimespan','Died','Age']]
df.head(10)
```

```python
df.loc[3]
```

```python
# меняем формат колонки born на формат to_datetime
df['Born']=df['Born'].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df['Born']=pd.to_datetime (df['Born'])
df.head(10)
```

```python
# разбиваем колонку Age atstart of presidency на две - Age at start of presidency, Date at start of presidency
pattern="(?P<Age_at_start_of_presidency>^[\w]{2}\s[\w]{5},\s[\w]{1,3}\s[\w]{4})(?P<Date_at_start_of_presidency>[\w]{3} [\w]{1,2}, [\w]{4}$)"
dates=df['Age atstart of presidency'].str.extract(pattern)
df['Age at start of presidency']= dates['Age_at_start_of_presidency']
df['Date at start of presidency']= dates['Date_at_start_of_presidency']
del df['Age atstart of presidency']
```

```python
# меняем формат колонки Date at start of presidency на формат to_datetime
df['Date at start of presidency']=pd.to_datetime (df['Date at start of presidency'])
df.head(10)
```

```python
df=df [['Last','First','Born','Age at start of presidency','Date at start of presidency','Age atend of presidency','Post-presidencytimespan','Died','Age']]
df['Date at start of presidency']=pd.to_datetime (df['Date at start of presidency'])
df.head(10)
```

```python
pattern="(?P<Age_at_end_of_presidency>^[\w]{2}\s[\w]{5},\s[\w]{1,3}\s[\w]{4})(?P<Date_at_end_of_presidency>[\w]{3} [\w]{1,2}, [\w]{4}$)"
dates=df['Age atend of presidency'].str.extract(pattern)
df['Age at end of presidency']= dates['Age_at_end_of_presidency']
df['Date at end of presidency']= dates['Date_at_end_of_presidency']
del df['Age atend of presidency']
df.head(10)
```

```python
df=df [['Last','First','Born','Age at start of presidency','Date at start of presidency','Age at end of presidency','Date at end of presidency','Post-presidencytimespan','Died','Age']]
df['Date at end of presidency']=pd.to_datetime (df['Date at end of presidency'])
df.head(10)
```
