```python
cylinders = set(d['cyl'] for d in mpg)
cylinders
```

```python
CtyMpgByCyl = []

for c in cylinders:
    summpg = 0
    cyltypecount = 0
    for d in mpg:
        if d['cyl'] == c:
            summpg += float(d['cty'])
            cyltypecount += 1
    CtyMpgByCyl.append((c, summpg / cyltypecount))
CtyMpgByCyl.sort(key=lambda x: x[0])
CtyMpgByCyl
```

```python
vehicleclass = set(d['class'] for d in mpg)
vehicleclass
```

```python
CylInClass=[]
for v in vehicleclass:
    cyltypecount = 0
    sumcyl = 0 
    for d in mpg:
        if d['class'] == v:
            sumcyl += float(d['cyl'])
            cyltypecount+=1
    CylInClass.append((v, sumcyl/cyltypecount))
CylInClass.sort(key=lambda x: x[0])
CylInClass
```

```python
grades='ACAAAABCBCBAA'
re.findall('BC', grades)
```

```python
re.findall('[CB]', grades) #в re квадатные скобки это оператор позволяющий найти
                            # все вхождения необходимых знаков (C и B)
```

```python
re.findall('[A][B-C]', grades) #ищем сочетания AC и AB
```

```python
re.findall('AB|AC', grades) # ищем АВ or(|) АС
```

```python
re.findall('[^A]', grades) #Ищем не А - '[^A]' внутри оператора SET - '^' это отрицание.  
```

```python
re.findall('^[^A]', grades) #Ищем в начале строки not А
```

```python
#Оператор "Квантификатор". e{m,n}, где е - символ или сочетание которое мы ищем, m - мин. кол-во вхождений, n - макс. кол-во вхождений.  
```

```python
re.findall('A{2,10}', grades) # шаблон ищет комбинацию от 2 до 10  А подряд. 
```

```python
re.findall('A{1,1}A{1,1}', grades) #шаблон ищет комбинацию двух А стоящих рядом. т.е. в ААА будет АА,АА
```

```python
re.findall('A{2, 2}', grades) #между m и n не должно быть пробелов. иначе получим []
```

```python
re.findall('A{2}', grades) #если одно значение в фигурных скобках, то оно и m и n
```

```python
re.findall('A{1,10}B{1,10}C{1,10}', grades)
```
